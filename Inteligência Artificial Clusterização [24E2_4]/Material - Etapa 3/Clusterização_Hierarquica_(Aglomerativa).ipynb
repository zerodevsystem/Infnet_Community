```python
import mglearn

mglearn.plots.plot_agglomerative_algorithm()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, _ = make_blobs(random_state=42)

hierarc = AgglomerativeClustering(n_clusters=3)
preds = hierarc.fit_predict(X)

preds
```

```python
mglearn.discrete_scatter(X[:, 0], X[:, 1], preds)
plt.legend(['Cluster 0', 'Cluster 1', 'Cluster 2'])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, ward, linkage

linkage_array = linkage(X)

dendrogram(linkage_array, truncate_mode='level', p=3)

ax = plt.gca()
bounds = ax.get_xbound()

ax.plot(bounds, [10, 10], '--', c='k')
ax.plot(bounds, [4, 4], '--', c='k')

ax.text(bounds[1], 10, 'dois clusters', va='center', fontdict={'size': 12})
ax.text(bounds[1], 4, 'três clusters', va='center', fontdict={'size': 12})

plt.xlabel("Índice da Amostra")
plt.ylabel("Distância")
plt.show()
```

```python
linkage_array = ward(X)

dendrogram(linkage_array, truncate_mode='level', p=3)

ax = plt.gca()
bounds = ax.get_xbound()

ax.plot(bounds, [80, 80], '--', c='k')
ax.plot(bounds, [40, 40], '--', c='k')

ax.text(bounds[1], 80, 'dois clusters', va='center', fontdict={'size': 12})
ax.text(bounds[1], 40, 'três clusters', va='center', fontdict={'size': 12})

plt.xlabel("Índice da Amostra")
plt.ylabel("Distância")
plt.show()
```

```python
X, _ = make_blobs(random_state=42)

hierarc = AgglomerativeClustering(n_clusters=None, distance_threshold=40, compute_full_tree=True)
preds = hierarc.fit_predict(X)

mglearn.discrete_scatter(X[:, 0], X[:, 1], preds)
cluster_range = list(range(preds.max() + 1))
plt.legend(['Cluster ' + str(k) for k in cluster_range])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
