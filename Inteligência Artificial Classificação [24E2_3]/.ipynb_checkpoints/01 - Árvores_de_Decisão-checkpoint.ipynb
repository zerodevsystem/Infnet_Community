# **Árvores de decisão**

```python
# Carregando  as bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

## Pima Indians Diabetes Dataset

Dados originalmente fornecidos pelo National Institute of Diabetes and Digestive and Kidney Diseases. O objetivo é predizer se um determinado paciente está propenso a ter diabetes ou não a partir de outros dados clínicos e laboratoriais.

Existem diversas restrições na seleção de observações. Este subconjunto dos dados contém apenas pacientes do sexo feminino com pelo menos 21 anos de idade e com descendência indígena Pima.

Variáveis:

- **Pregnancies**: Number of times pregnant
- **Glucose**: Plasma glucose concentration a 2 hours in an oral glucose tolerance test
- **BloodPressure**: Diastolic blood pressure (mm Hg)
- **SkinThickness**: Triceps skin fold thickness (mm)
- **Insulin**: 2-Hour serum insulin (mu U/ml)
- **BMI**: Body mass index (weight in kg/(height in m)^2)
- **DiabetesPedigreeFunction**: Diabetes pedigree function
- **Age**: Age (years)
- **Outcome**: Class variable (0 or 1)

```python
# Carregando a base de dados
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
pima = pd.read_csv('pima-indians-diabetes.csv')
pima.columns = col_names
```

```python
pima
```

Avaliando o balanceamento da base

```python
# Visualizando a distribuição da variável-alvo, em percentuais
ax = sns.countplot(pima, x='label', stat="percent")
ax.bar_label(ax.containers[0], fontsize=10);
```

```python
# Separando o dataset em variáveis independentes e variável-alvo
feature_cols = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']
X = pima[feature_cols] # Variáveis independentes
y = pima.label # Variável-alvo
```

```python
# Dividindo a base em conjuntos de Treino e Teste (70% Treino and 30% Teste)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Instanciando, treinando o modelo e aplicando o resultado na base de testes

# Instanciando o classificador
clf = DecisionTreeClassifier()

# Realizando o treino
clf = clf.fit(X_train,y_train)

# Aplicando o modelo treinado no conjunto de Teste
y_pred = pd.DataFrame(clf.predict(X_test))
```

```python
# Avaliando os resultados

# Matriz de confusão
# Linha: Real
# Coluna: Predito
print(confusion_matrix(y_test, y_pred))
```

```python
# Classification Report
print(classification_report(y_test, y_pred))
```

```python
# Visualização da árvore gerada.
from sklearn import tree

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (50,50), dpi=300)
tree.plot_tree(clf, filled=True)
fig.savefig('tree_sklearn.png')
```

```python
# Otimizando o modelo através o parâmetro "max_depth".

# Criando um novo objeto classificador
clf = DecisionTreeClassifier(max_depth=5)

# Treinando novamente o modelo
clf = clf.fit(X_train,y_train)

# Aplicando o modelo no conjunto de Teste
y_pred = pd.DataFrame(clf.predict(X_test))
```

```python
 # Avaliando o novo resultado

# Matriz de confusão
# Linha: Real
# Coluna: Predito
print(confusion_matrix(y_test, y_pred))
```

```python
# Classification Report
print(classification_report(y_test, y_pred))
```

```python
# Visualizando a nova árvore gerada.
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (15,15), dpi=300)
tree.plot_tree(clf, filled=True)
fig.savefig('tree_sklearn_2.png')
```

# **Usando o GridSearch na busca dos melhores parâmetros**



```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import GridSearchCV

# Instanciando o classificador
clf = DecisionTreeClassifier()

tree_para = {'criterion':['gini','entropy'],
             'max_depth':[4,5,6,7,8,9,10,11,12,15,20,30,40,50,70]}
clf_gs = GridSearchCV(DecisionTreeClassifier(), tree_para, cv=5, refit=True)

# Realizando o treino
clf_gs.fit(X, y)
```

# **Executando uma Árvore de Decisão em outra base**

Utilizaremos agora uma base de dados relacionada à processos de anti-fraude.

Apesar dos dados estarem anonimizados, por motivos de proteção de dados pessoais, eles mantêm as características matemáticas dos dados originais de forma que possam ser utilizados em processos de modelagem.

A anonimizaçao dos dados é um processo obrigatório em muitas áreas de negócio.

```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
# Carregando a base
base_treinamento = pd.read_csv('fraud_dataset.csv')
```

```python
base_treinamento
```

```python
# Fazendo a separação dos dados (variáveis independente e dependente)

# Variáveis independentes
X = base_treinamento[[
    'V1',
    'V2',
    'V3',
    'V4',
    'V5',
    'V6',
    'V7',
    'V8',
    'V9',
    'V10',
    'V11',
    'V12',
    'V13',
    'V14',
    'V15',
    'V16',
    'V17',
    'V18',
    'V19',
]]

# Variável-alvo
y = base_treinamento['Class']
```

```python
# Garantindo que não existem NAs para não termos problemas na modelagem.
X = X.fillna(0)
```

```python
# Transformando tudo para float
X = X.astype(float)
```

```python
# Visualizando a distribuição da variável-alvo, em percentuais
ax = sns.countplot(base_treinamento, x='Class', stat="percent")
ax.bar_label(ax.containers[0], fontsize=10);
```

```python
# Separando dados para Treino e Teste

# Separando os dados de treino e teste para o modelo
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
# Aplicando um modelo de Árvore de Decisão.
clf = DecisionTreeClassifier(max_depth=5)
clf.fit(X_train,y_train)

# Aplicando o modelo na base de Teste
y_pred_teste = clf.predict(X_test)
```

```python
# Avaliando o resultado
cm = metrics.confusion_matrix(y_test, y_pred_teste)

disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
```

```python
print(metrics.classification_report(y_test, y_pred_teste,
                                    target_names=['Saudavel','Suspeito']))
```

```python
# Criando uma Curva ROC.
from sklearn.metrics import RocCurveDisplay

display = RocCurveDisplay.from_estimator(
        clf,
        X_test,
        y_test,
        name="Curva ROC",
        color="darkorange",
    )
_ = display.ax_.set(
    xlabel="False Positive Rate",
    ylabel="True Positive Rate",
    title="Curva ROC",
)
```

```python
# Visualização da árvore gerada.
from sklearn import tree

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (50,50), dpi=300)
tree.plot_tree(clf, filled=True)
fig.savefig('tree2_sklearn.png')
```
